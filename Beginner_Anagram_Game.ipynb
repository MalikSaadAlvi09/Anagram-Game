## Anagram Game For Beginner Python Project

```python
import random
words = ["ironman", "thor", "hawkeye", "wanda", "vision"]
word = random.choice(words)
#print(word)
jumble = "".join(random.sample(word, len(word)))
#print(jumble)
chances = 3
print("~"*30)
print("~~~ Avengers Jumble Words ~~~")
while chances != 0:
    print("The word is: ", jumble)
    guess = input("Enter your guessed word: ").lower()
    if guess == word:
        print("Correct Guess.")
        print("You Won!")
        break
    else:
        chances -= 1
        print("Incorrect Guess!!")
        print("Remaining chances are: ", chances)
else:
    print("All your chances are exhausted.")
    print("You loss.")
    print("The correct word is: ", word)
print('Thanks For Playing Our game')
```

## To Restart Game by Using Loop

```python
import random

while True:

    f = open("Random_Words.txt", "r")
    a = f.readline()
    words = a.split(",")

    word = random.choice(words).strip()

    jumble = "".join(random.sample(word, len(word)))

    chances = 3

    print("~" * 30)
    print("~~~ Avengers Jumble Words ~~~")

    while chances != 0:

        print("The word is:", jumble)

        guess = input("Enter your guessed word: ").lower()

        if guess == word:
            print("Correct Guess.")
            print("You Won!")
            break

        else:
            chances -= 1
            print("Incorrect Guess!!")
            print("Remaining chances are:", chances)

    else:
        print("All your chances are exhausted.")
        print("You lost.")
        print("The correct word is:", word)

    print("Thanks For Playing Our Game")

    # Restart option
    play_again = input("Do you want to play again? (yes/no): ").lower()

    if play_again != "yes":
        print("Game Ended.")
        break
```
